```python
#Task A

def blockmultiply(A,B,block_size):
    result = np.zeros((A.shape[0],B.shape[1]))
    for i in range(0,A.shape[0],block_size):
        for j in range(0,B.shape[1],block_size):
            for k in range(0, A.shape[1],block_size):
                result[i:i+block_size,j:j+block_size]+=np.dot(A[i:i+block_size,k:k+block_size],B[k:k+block_size,j:j+block_size])
    return result
A = np.array([[1,2,3,4,6,7,8,3], [5,6,7,8,6,5,7,8],[9,10,11,12,6,7,8,5],[13,14,15,16,5,6,7,2],[9,8,6,7,5,4,3,2],[8,9,3,5,5,6,7,3],[3,4,5,2,1,1,2,3],[5,4,3,7,6,9,0,0]])
B = np.array([[16,15,14,13,10,10,20,0],[12,11,10,9,7,7,6,6],[8,7,6,5,5,5,6,6],[4,3,2,1,7,7,7,7],[5,6,7,8,10,10,10,10],[12,11,10,9,0,0,9,0],[1,2,3,4,5,5,6,6],[7,8,9,6,9,0,8,0]])
block_size = 4
result=blockmultiply(A,B,block_size)
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResultant Matrix:")
print(result)
```

```python
#Task C
from bs4 import BeautifulSoup
import requests


for n in range(1, 10):
    url = "https://www.amazon.in/s?k=iPhone"

    response = requests.get(url)
    if (response.status_code == 200):
        print("OK", n)
        soup = BeautifulSoup(response.text, "html.parser")
        data = []
        for product in soup.find_all(string = re.compile("iPhone")):
            title = product.text.strip()
            print(title)
            data.append([title])
        with open("amazon.csv", "a", encoding = "utf-8") as csvfile:
                csvwriter = csv.writer(csvfile)
                csvwriter.writerows(data)
```
